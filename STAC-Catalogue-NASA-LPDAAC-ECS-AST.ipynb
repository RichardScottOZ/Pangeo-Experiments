```python
import copy
stac_new = copy.deepcopy(stac_items)
```

The asset band names are unique, so we don't want to have a list of 86 different band names but to combine by each, so we will remap them below to something common - e.g. 01, 02 could be V and T, etc.

```python
for index, i in enumerate(stac_items):
    #print(i.assets)
    for k in i.assets:
        if '_T' in k:
            stac_new[index].assets['B01'] = stac_items[index].assets[k]
        if '_V' in k:
            stac_new[index].assets['B02'] = stac_items[index].assets[k]
```

```python
stack = stackstac.stack(stac_new, epsg=6933, resolution=90, resampling=1, assets=['B01'])
#stack = stackstac.stack(stac_items, epsg=6933, resolution=30, resampling=1)

stack
```
