```python
import requests
import pandas as pd

import intake
limit = 500

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
nasa = 'https://cmr.earthdata.nasa.gov/stac/'

response = requests.get(nasa)
json_response = response.json()
```

```python
json_response = response.json()
```

```python
#{'title': 'LARC_ASDC', 'rel': 'child', 'type': 'application/json', 'href': 'https://cmr.earthdata.nasa.gov/stac/LARC_ASDC'}
collection_list = []
for collection in json_response["links"]:
    collection_list.append(collection)
```

```python
datasets_list = []
for datasets in collection_list:
    url = datasets["href"]
    response = requests.get(url)
    json_response = response.json()
    datasets_list.append(json_response)
```

```python
#need to handling paging when more than one, this just a quick look to narrow down which collection ASTER in
collection_list = []
dataset_list = []
for dataset in datasets_list:
    df = pd.DataFrame([dataset])
    collection_list.append(df)
    
    for link in dataset["links"]:
        ldf = pd.DataFrame([link])
        ldf["id"] = dataset["id"]
        ldf = ldf.loc[ldf["rel"] == "child"]
        dataset_list.append(ldf)
    
dfCollections = pd.concat(collection_list)    

dfDataset = pd.concat(dataset_list)    
```
